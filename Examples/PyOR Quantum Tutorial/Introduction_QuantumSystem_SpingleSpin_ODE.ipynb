### Schr√∂dinger Equation

```python
# Initial State
vec = QLib.Basis_Ket(2,0)
```

```python
vec.matrix
```

```python
# Rotate by 90 deg about X axis
vec = vec.Rotate(90,QS.Ix)
```

```python
vec.Round(3).matrix
```

```python
# Hamiltonian
Hz = 2.0 * np.pi * 10 * QS.Iz
```

```python
Hz.matrix
```

```python
# Evolution
QLib.AcqAQ = 1
QLib.AcqDT = 0.001
t, vect = QLib.Evolve_SE_ODE(vec,Hz)
```

```python
# Expectation
t, signal_SE = QLib.Expectation(t,vect,QS.Ix)
```

```python
# Plotting
plt.figure(1)
plt.plot(t,signal_SE)
```

### Liouville Equation : Hilbert Space

```python
# Initial density matrix
rho = QS.Iz
```

```python
rho.matrix
```

```python
# Rotate about x axis by 90 deg
rho = rho.Rotate(90,QS.Ix)
```

```python
# Hamiltonian
Hz = 2.0 * np.pi * 10 * QS.Iz
```

```python
# Evolution
QLib.AcqAQ = 1
QLib.AcqDT = 0.001
t, rhot = QLib.Evolve_Hilbert_ODE(rho,Hz)
```

### Liouville Equation : Liouville Space

```python
# Initial density matrix
rhoL = QS.Iz
```

```python
rhoL.matrix
```

```python
# Vectorize
#Lib.RowColOrder = 'C' # Vectorize by row
QLib.RowColOrder = 'F' # Vectorize by col
Lrho = QLib.DMToVec(rhoL)
```

```python
Lrho.matrix
```

```python
# Rotate by 90 deg about X axis
Lrho = Lrho.Rotate(90,QLib.CommutationSuperoperator(QS.Ix))
```

```python
Lrho.matrix
```

```python
# Hamiltonian
Hz = 2.0 * np.pi * 10 * QS.Iz
HzL = QLib.CommutationSuperoperator(Hz)
```

```python
Hz.matrix
```

```python
HzL.matrix
```

```python
# Evolution
QLib.AcqAQ = 1
QLib.AcqDT = 0.001
t, Lrhot = QLib.Evolve_Liouville_ODE(Lrho,HzL)
```

```python
# Expectation
t, signal_Li = QLib.Expectation(t,Lrhot,QLib.DMToVec(QS.Ix).Adjoint())
```

```python
# Plotting
plt.figure(3)
plt.plot(t,signal_Li)
```
